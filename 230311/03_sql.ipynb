```python
## 라이브러리 로드
import pymysql
import pandas as pd
```

```python
## 서버 정보와 서버의 접속
_db = pymysql.connect(
    host = '192.168.0.90',
    user = 'ezen',
    password = '1234',
    db = 'test',
    port = 3306
)
```

```python
## cursor 생성
## sql 쿼리문을 실행 시키기 위해 만들어야되는 필수 변수
## DicCursor : 쿼리문이 실행되고 나온 결과물을 딕셔너리 형태로 받아온다
cursor = _db.cursor(pymysql.cursors.DictCursor)
```

```python
sql = """ 
    SELECT
    *
    FROM
    corona
"""

cursor.execute(sql)

result = cursor.fetchall()
print(result)
```

```python
pd.DataFrame(result)
```

```python
## insert 문 
sql = """
    INSERT INTO
    user
    VALUES(
    'test',
    '1234',
    'test',
    0
    )
"""

cursor.execute(sql)
```

```python
_db.commit()
```

```python
_id = input('id를 입력하시오')
_pass = input('password를 입력하시오')
_name = input('name을 입력하시오')
_gender = int(input('gender을 입력하시오'))

sql = """
    INSERT INTO
    user
    VALUES
    (%s, %s, %s, %s)
"""

sql2 = f"""
    INSERT INTO
    user
    VALUES
    ('{_id}', '{_pass}', {_name}', {_gender}')
"""

cursor.execute(sql, [_id, _pass, _name, _gender])
## cursor.execute(sql2)
_db.commit()
```

```python
## update 문 

sql = """
    UPDATE 
    user
    SET
    `password` = %s
    WHERE
    `id` = %s
"""

_value = input()
_id = input()

cursor.execute(sql, [_value, _id])
_db.commit()
```
