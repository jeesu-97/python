### Matplotlib (B6)

데이터를 차트나 플롯으로
    * plot() -> 라인 그래프
        * 옵션 부분에 색상인자 값이나 헥스 값 넣어주면 선의 색 변경
        * 옵션 부분에 선 종류 인자 값을 넣어주면 선의 종류 변경

        * plt.xlabel()
        * plt.ylabel()
        * plt.title()
        * plt.legend() -> 그래프 라인의 범례를 추가 (loc = "위치" 지정하면 범례 위치 고정 가능)

    * subplot(nrow, ncol, pos) 
        * 여러개의 그래프 그릴 때 사용
        * subplot(행의 수, 열의 수, 위치)

    * bar() -> 막대 그래프 (x값, y값 모두 지정해야 함)
        * xticks() 이용하여 x축의 레이블 값 변경 가능
        * 색상 (color), 그래프 폭 (width) 변경 가능

    * scatter() -> 산점도
        * 점 하나하나의 크기 및 색상 지정 가능

```python
import platform
```

```python
x = [1, 2, 3, 4]
y1 = [3, 6, 2, 7]
y2 = [1, 10, 3, 1]

## 한글이 깨지는 현상을 막기 위해 폰트 변경
if platform.system() == 'Darwin': ## Mac 인 경우
    plt.rc('font', family='AppleGotie')  
else :
    plt.rc("font", family='Malgun Gothic')

## 선의 색
plt.plot(x, y1, "r")

## 선의 종류
plt.plot(x, y2, "--")

## x축의 이름
plt.xlabel("Class")

## y축의 이름
plt.ylabel("Count")

## title
plt.title("그래프")

## 범ㄹ메 추가
plt.legend(["A", "B"])

plt.show()
```

```python
## subplot(행의 수, 열의 수, 위치) (B16)

## 그래프 2개 그린다는 가정

## subplot을 이용해서 위치를 지정
plt.subplot(2, 1, 1)
plt.plot(x, y1)

plt.subplot(2, 1, 2)
plt.plot(x, y2)

plt.show()
```

```python
## 바형 그래프 (B17)
## x축 값 필수로 지정해야 함

x = [1, 2, 3, 4]
y = [80, 60, 90, 70]
```

```python
plt.bar(x, y)
plt.show
```

```python
## 막대의 너비와 색상 변경
plt.bar(x, y, color=['g', 'r'], width=0.5)

## x축의 이름을 변경
plt.xticks(x, ['maths', 'science', 'english', 'history'])

plt.show()
```

```python
## 산점도 그래프 (B18)

import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
## 영역 (area), 색 (color), 투명도 (alpha) 변경
color = np.random.rand(50)
area = (20 * np.random.rand(50)) ** 2 ## 굉장히 작으므로 곱, 제곱 해주는 것

plt.scatter(x, y, s = area, c = color, alpha = 0.5)
plt.show()
```
